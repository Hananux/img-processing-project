```python

img = cv2.imread("tf2_engineer.jpg")


img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)


plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
# Get image dimensions
height, width, _ = img.shape

# Compute center coordinates
y_c = height // 2
x_c = width // 2

# Print the center coordinates
print(f"Image center: (y={y_c}, x={x_c})")

# Get BGR intensity at center
center_bgr = img[y_c, x_c]
print(f"Intensity at center (BGR): {center_bgr}")
```

```python
# Make a copy of the image to draw on
img_patch = img.copy()

# Patch size
patch_height = 30  # pixels
patch_width = 40   # pixels

# Calculate rectangle coordinates
top_left = (x_c - patch_width // 2, y_c - patch_height // 2)
bottom_right = (x_c + patch_width // 2, y_c + patch_height // 2)

# Patch color: #329ea8 → BGR = (168, 158, 50)
patch_color_bgr = (168, 158, 50)

# Draw filled rectangle on the image
cv2.rectangle(img_patch, top_left, bottom_right, patch_color_bgr, thickness=-1)
```

```python
# Check the center pixel again
patched_center_bgr = img_patch[y_c, x_c]
print(f"Intensity at center after patch (BGR): {patched_center_bgr}")

# Convert to RGB for display
img_patch_rgb = cv2.cvtColor(img_patch, cv2.COLOR_BGR2RGB)

# Show the patched image
plt.imshow(img_patch_rgb)
plt.title("Image with Color Patch")
plt.axis("off")
plt.show()
```

```python
#first load the img
import numpy as np

img = cv2.imread("pout.tif", cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
#2ndd log transform
img_float = img.astype(np.float32)
log_img = np.log1p(img_float)
log_norm = cv2.normalize(log_img, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)

plt.imshow(log_norm, cmap='gray')
plt.title("Log Transform")
plt.axis("off")
plt.show()
```

```python
# 3rd! Inverse log of og img

img_float_scaled = img.astype(np.float32) / 255.0  # scale range before applying cuz it showed black
inv_log = np.expm1(img_float_scaled * 5)  # adjust scaling factor
inv_log_norm = cv2.normalize(inv_log, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)

plt.imshow(inv_log_norm, cmap='gray')
plt.title("Inverse Log (Original, Scaled)")
plt.axis("off")
plt.show()

```

```python
# 4 Inverse of log transformed image
inv_of_log = np.expm1(log_img)
inv_of_log_norm = cv2.normalize(inv_of_log, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)

plt.imshow(inv_of_log_norm, cmap='gray')
plt.title("Inverse of Log-Transformed")
plt.axis("off")
plt.show()
```

```python
#Original vs Log Transform
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title("Original")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(log_norm, cmap='gray')
plt.title("Log Transform")
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
#1. Read and display the original image in grayscale.
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load img
img = cv2.imread("moon.tif", cv2.IMREAD_GRAYSCALE)

# original image
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
#2 Apply unsharp masking in spatial domain using le formula'
blurred = cv2.GaussianBlur(img, (5, 5), 0)
mask = img.astype(np.float32) - blurred.astype(np.float32)

k_values = [0.2, 0.5, 1.0]
spatial_results = []

for k in k_values:
    sharpened = img.astype(np.float32) + k * mask
    sharpened = np.clip(sharpened, 0, 255).astype(np.uint8)
    spatial_results.append(sharpened)
```

```python
# High-pass filter function
def high_pass_filter(shape, D0):
    rows, cols = shape
    crow, ccol = rows // 2, cols // 2
    H = np.zeros((rows, cols), dtype=np.float32)
    for u in range(rows):
        for v in range(cols):
            D = np.sqrt((u - crow)**2 + (v - ccol)**2)
            H[u, v] = 1 if D >= D0 else 0
    return H

# Prepare FFT of the image
f_shift = np.fft.fftshift(np.fft.fft2(img))
H = high_pass_filter(img.shape, D0=30)

freq_results = []
for k in k_values:
    G = (1 + k * H) * f_shift
    g = np.fft.ifft2(np.fft.ifftshift(G)).real
    g = np.clip(g, 0, 255).astype(np.uint8)
    freq_results.append(g)
```

```python
# Display all sharpened results for comparison
plt.figure(figsize=(12, 8))
for i in range(3):
    # Spatial domain
    plt.subplot(2, 3, i + 1)
    plt.imshow(spatial_results[i], cmap='gray')
    plt.title(f"Spatial k={k_values[i]}")
    plt.axis("off")

    # Frequency domain
    plt.subplot(2, 3, i + 4)
    plt.imshow(freq_results[i], cmap='gray')
    plt.title(f"Freq k={k_values[i]}")
    plt.axis("off")

plt.tight_layout()
plt.show()
```
